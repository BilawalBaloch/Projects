```python
# Convert relevant columns to numeric type if they are not already
data['Sales'] = pd.to_numeric(data['Sales'], errors='coerce')
data['Profit'] = pd.to_numeric(data['Profit'], errors='coerce')
data['Discount'] = pd.to_numeric(data['Discount'], errors='coerce')
data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce')

plt.plot(data['Sales'], data['Profit'], data['Discount'], data['Quantity'])
plt.legend()
plt.show()
```

```python

plt.subplot(3, 2, 4)
plt.bar(['Quantity'], data['Quantity'].mean(), color='blue', label='Quantity')
plt.bar(['Discount'], data['Discount'].mean(), color='yellow', label='Discount')
plt.title('Quantity vs Discount')
plt.ylabel('Average Value')
plt.legend()
```

```python

plt.subplot(3, 2, 6)
plt.bar(['Discount'], data['Discount'].mean(), color='blue', label='Discount')
plt.bar(['Profit'], data['Profit'].mean(), color='yellow', label='Profit')
plt.title('Discount vs Profit')
plt.ylabel('Average Value')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(3, 5))
sns.barplot(x=['Profit', 'Quantity'], y=[data['Profit'].mean(), data['Quantity'].mean()])
plt.title('Average Profit, Sales')
plt.xlabel('Metric')
plt.ylabel('Average Value')

# Show the plot
plt.show()
```

```python

data['Order Date'] = pd.to_datetime(data['Order Date'])
sales_by_date = data.groupby('Order Date')['Sales'].sum()
sales_by_date.plot(figsize=(12, 6))
plt.title('Daily Sales Trend')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
