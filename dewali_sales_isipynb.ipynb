```python
plt.figure(figsize=(5,3))
axis = sns.countplot(data=df, x="Gender" , hue="Gender")
for bar in axis.containers:
    axis.bar_label(bar)
```

```python
plt.figure(figsize=(10,7))
c = {'26-35':'green','36-45':'blue','18-25':'yellow','46-50':'pink','51-55':'red','55+':'orange','0-17':'green'}
data = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=data, x="Age Group", y="Amount", palette=c)
```

```python
plt.figure(figsize=(10,7))
sns.countplot(data=df, x="Marital_Status")
c = {'Himachal Pradesh':'green', 'Bihar':'blue', 'Haryana':'yellow', 'Delhi':'pink', 'Gujarat':'red',
     'Uttar Pradesh':'orange', 'Karnataka':'purple', 'Andhra\xa0Pradesh':'brown', 'Madhya Pradesh':'gray',
     'Maharashtra':'cyan'}
data = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=data, x="State", y="Amount", palette=c)
```

```python
df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
plt.figure(figsize=(10,7))
c = {'Himachal Pradesh':'green', 'Bihar':'blue', 'Haryana':'yellow', 'Delhi':'pink', 'Gujarat':'red',
     'Uttar Pradesh':'orange', 'Karnataka':'purple', 'Andhra\xa0Pradesh':'brown', 'Madhya Pradesh':'gray',
    'Maharashtra':'cyan'}
data = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)  # Removed extra space before 'Orders'
unique_states = data['State'].unique()
colors = sns.color_palette("husl", len(unique_states))
c = dict(zip(unique_states, colors))
sns.barplot(data=data, x="State", y="Orders", palette=c)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
