## **ML model from sklearn**

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X = data_1.drop(['prcp','tmax','tmin', 'station', 'name'], axis=1)
```

```python
y = data_1.drop(['station', 'name', 'year_feature'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size=0.3, random_state=42)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pre = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pre)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pre)
```

```python
print(f"\nModel Evaluation:")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R2) Score: {r2:.4f}")
```

```python
y_test_df = pd.DataFrame(y_test, index=X_test.index)
y_pre_df = pd.DataFrame(y_pre, index=X_test.index, columns=y_test.columns)

combined_df = pd.concat([y_test_df, y_pre_df], axis=1)
combined_df.columns = ['y_test_prcp', 'y_test_tmax', 'y_test_tmin', 'y_pre_prcp', 'y_pre_tmax', 'y_pre_tmin']

combined_df['diff_prcp'] = np.abs(combined_df['y_test_prcp'] - combined_df['y_pre_prcp'])
combined_df['diff_tmax'] = np.abs(combined_df['y_test_tmax'] - combined_df['y_pre_tmax'])
combined_df['diff_tmin'] = np.abs(combined_df['y_test_tmin'] - combined_df['y_pre_tmin'])

combined_df.sort_values("diff_tmax", ascending=False)
```
